# Daftar Kontak

```python
# Pilihan Menu
def menu():
    print(" --Menu-- ")
    print("\t 1. Daftar Kontak")
    print("\t 2. Tambahkan Kontak")
    print("\t 3. Keluar")
    
print(" Selamat Datang! ")
# menyiapkan penyimpanan nama dan no hp
nama = ['Budi','Ani']
no_hp = ['0898767865','07896543567']
# memanggil pilihan menu
while True:
    print('')
    menu()
    pilih = input("Pilih Menu : ")
    if pilih == "1":
        print("Daftar kontak :")
        for i in range(len(nama)):
            print("Nama  : ",nama[i])
            print("No HP : ",no_hp[i])
            print("---------------------------")
    elif pilih == "2":
        print("Masukan kontak baru :")
        name = input("\t Nama :")
        phone = input("\t No Hp :")
        nama.append(name)
        no_hp.append(phone)
        print("\t kontak berhasil ditambahkan!")
        print("-------------------------------")
        
    elif pilih == "3":
        print("Program selesai, sampai jumpa!")
        print("------------------------------")
        break
    else:
        print("\nMaaf, menu tidak tersedia")
        print("-------------------------------")
```

# Soal Bonus Kalkulator & Chatbot

```python

## Fungsi Menu awal
def menu():
    print("Silahkan pilih bantuan yang bisa aku berikan:")
    print("1. Kalkulator sederhana")
    print("2. Menemani Curhatmu")
    print("3. Keluar aplikasi")
    
# Fungsi Kalkulator
def kalkulator():
    while True:
        print("\nAplikasi Kalkulator sederhana :")
        print("--------------------------------------")
        print("Ayo pilih menu :")
        print("a.Penjumlahan")
        print("b.Pengurangan")
        print("c.Perkalian")
        print("d.Pembagian")
        print("e.Kembali")
        operasi = input(" Pilih hurufnya saja :")
        if operasi =="a":
            x = int(input("Masukan angka pertama:"))
            y = int(input("Masukan angka kedua:"))
            s = x+y
            print ("Maka hasil dari",x,"+",y,"adalah",s)
        elif operasi =="b":
            x = int(input("Masukan angka pertama:"))
            y = int(input("Masukan angka kedua:"))
            k = x-y
            print("Maka hasil dari",x,"-",y,"adalah",k)
        elif operasi == "c":
            x = int(input("Masukan angka pertama:"))
            y = int(input("Masukan angka kedua"))
            t = x*y
            print("Maka hasil dari",x,"x",y,"adalah",t)
        elif operasi == "d":
            x = int(input("Masukan angka pertama:"))
            y = int(input("Masukan angka kedua:"))
            g = x/y
            print("Maka hasil dari",x,":",y,"adalah",g)
        elif operasi =="e":
            print("\nTerima kasih untuk pemakaian kalkulator ini!")
            print("--------------------")
            break
        else:
            print("\nMaaf, pilihanmu tidak tersedia")
            print("--------------------")
        
# Fungsi Chatbot        
def chatbot():
    input("Coba ceritakan apa yang kamu rasakan saat ini:")
    while True:
        cerita= input("\nCeritamu menarik, Coba ceritakan lebih banyak :")
        if 'gagal'in cerita or 'k' in cerita or 'bodoh' in cerita or 'jelek' in cerita:
            print("\n\tAyo lebih semangat lagi, kebaikan menunggumu!")
        elif 'kecewa' in cerita or 'sedih' in cerita or 'putus asa' in cerita:
            print("\n\tApa yang membuat kamu seperti ini? :")
        elif 'senang' in cerita or 'berhasil' in cerita or 'menang' in cerita or 'hadiah' in cerita:
            print('Itu sangat menarik!')
        else:
            print('\n\tTerima kasih sudah bersedia untuk bercerita! Semoga perasaanmu menjadi lebih baik.')
            print("----------------------------------")
            break
# Pengantar 
print("Hai, aku AIs, asisten virtual pertamamu ")
print("Siapa namamu?")

nama = input(" Masukkan nama : ")

print("\n Salam kenal",nama)
print("Apakah ada yang bisa aku bantu?")

# Munculkan menu awal
while True:
    print('')
    menu()
    
    pilih = input(" Pilih angka yang kamu butuhkan : ")
    # Menu Kalkulator
    if pilih =='1':
        kalkulator()
    elif pilih =='2':
        chatbot()
    elif pilih =='3':
        print("Sampai berjumpa lagi")
        print("----------------------------")
        break
    else:
        print("\nMaaf, angka yang kamu pilih tidak tersedia")
        print("----------------------------")
```
